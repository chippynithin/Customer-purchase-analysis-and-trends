#### CREATING DERIVED COLUMNS

Adding extra columns for Age group, Spender Category, and customer type by grouping them into different categories.

```python
conditions=[0,25,40,60,100]
labels=["Gen Z","Millenials","Gen X", "Boomer"]
df["Age_Group"]=pd.cut(df["Age"],bins=conditions,labels=labels)
```

```python
A_median=df["Amount_USD"].median()
bins=[0,A_median,100]
labels=["Low Spender","High Spender"]
df["Spender Category"]=pd.cut(df["Amount_USD"],bins=bins,labels=labels)
```

```python
bins=[0,1,5,20,40,50]
labels=["New","Occassional","Repeat","Loyal","VIP"]
df["Customer Type"]=pd.cut(df["Previous Purchases"],bins=bins,labels=labels)
```

#### 7.1 PIE CHART

Pie chart is used to visualise the part-to-whole relationship. It expresses the numerical ratio of parts of the whole in a variable as slices of a pie.

```python
age=df["Age_Group"].value_counts()
gender=df["Gender"].value_counts()
pay=df["Payment Method"].value_counts()
ship=df["Shipping Type"].value_counts()
```

```python
plt.figure(figsize=(15,10))
plt.subplot(2,2,1)
explode=[0.1,0,0,0]
plt.pie(age.values,labels=age.index,autopct='%1.1f%%',explode=explode,shadow= True,startangle=90)
plt.title("Age group")
plt.subplot(2,2,2)
explode=[0.1,0]
plt.pie(gender.values,labels=gender.index,autopct='%1.1f%%',explode=explode,shadow= True,startangle=90)
plt.title("Gender")
plt.subplot(2,2,3)
plt.pie(pay.values,labels=pay.index,autopct='%1.1f%%',shadow= True,startangle=90)
plt.title("Payment Type")
plt.subplot(2,2,4)
plt.pie(ship.values,labels=ship.index,autopct='%1.1f%%',shadow= True,startangle=90)
plt.title("Shipping Type")
plt.show()
```

#### 7.4 STACKED BAR CHARTS

Stacked bar charts are used to display the distribution of multiple variables within a single category.

The following stacked bar chart is used to analyse the colour preference in each category. I have taken top 5 colours for the analysis.

```python
c=df.groupby(["Category","Color"]).size().reset_index(name="count")
#c
```

```python
d=c.sort_values(["Category","count"],ascending=[True,False])
#d
```

```python
top5_col=d.groupby("Category").head(5)
#top5_col
```

```python
table_c=top5_col.pivot(index="Category",columns="Color",values="count").fillna(0)
table_c
```

```python
color_map={"Black":"black",
           "Blue":"blue",
           "Cyan":"cyan",
           "Gray":"Gray",
           "Lavender":"#E6E6FA",
           "Magenta":"magenta",
           "Maroon":"maroon",
           "Olive":"olive",
           "Peach":"#FFDAB9",
          "Pink":"pink",
          "Silver":"silver",
          "Teal":"teal",
          "Violet":"violet",
          "Yellow":"Yellow"}
plot_colors=[color_map.get(color,'white') for color in table_c.columns]

table_c.plot(kind="bar",stacked=True,color=plot_colors)
plt.legend(bbox_to_anchor=(1.05,1))
plt.xticks(rotation=45)
plt.title("Colour preference in each category")
plt.ylabel("Count")
plt.show()
```
